## Importing required libraries

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import Dropout
```

## Setting seeds for re producing same results each time

```python
np.random.seed(10)
```

## importing dataset

```python
data = pd.read_csv(R"F:\Data Science\Assignments\Neural Network\50_Startups (2).csv")
data.head()
```

## creating dummies


```python
data=pd.get_dummies(data,columns=['State'])
data.shape
```

## seperating target and input data

```python
x=data.drop("Profit",axis=1)
y=data.Profit
```

## splitting data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state =1)
x_train.shape
from sklearn.metrics import mean_squared_error
```

## defining model

```python
def design_mlp():
    # Initializing the model 
    model = Sequential()
    model.add(Dense(15,input_dim =6,activation="relu"))
    model.add(Dense(7,activation="relu"))
    model.add(Dense(1,activation="linear"))
    model.compile(loss="mean_squared_error",optimizer="adam",metrics=["RootMeanSquaredError"])
    return model

model = design_mlp()
```

```python
## fitting model
```

```python
model.fit(x=x_train,y=y_train,batch_size=1,epochs=1000)
test_accu=model.predict(x_test)
test_accu.shape

```

```python
model.evaluate(x_test,y_test)
```
